```python
!pip install openai
import pandas as pd
import concurrent.futures
from openai import OpenAI

# OpenAI 클라이언트 초기화
client = OpenAI(
    api_key="...",  # 여기에 실제 API 키를 입력하세요.
)

def summarize_case_law(case_text, max_length=4000):
    """
    주어진 판례 내용을 요약하는 함수입니다. 텍스트가 너무 길 경우, 최대 길이에 맞춰 자릅니다.

    :param case_text: 요약할 판례의 전문 또는 주요 내용
    :param max_length: 모델에 전달할 최대 텍스트 길이
    :return: 요약된 내용
    """
    try:
        # 텍스트가 최대 길이를 초과하는 경우, 앞부분만 잘라서 사용
        if len(case_text) > max_length:
            case_text = case_text[:max_length]

        prompt = f"다음은 한 판례의 내용입니다:\n\n{case_text}\n\n이 판례를 법률 전문가의 시각에서 요약해 주세요. 중요한 법적 이슈와 결론에 집중하세요."

        response = client.chat.completions.create(
            messages=[
                {
                    "role": "system",
                    "content": prompt,
                }
            ],
            model="gpt-3.5-turbo",  # 사용할 모델 지정
        )

        # 모델의 응답 반환
        return response.choices[0].message.content
    except Exception as e:
        print(f"오류 발생: {e}")
        return None

def summarize_cases_from_csv(csv_file_path):
    """
    CSV 파일에서 판례 내용을 읽어와 요약합니다.

    :param csv_file_path: 판례 내용이 포함된 CSV 파일의 경로
    :return: 판례일련번호와 각 판례에 대한 요약 목록
    """
    try:
        df = pd.read_csv(csv_file_path, encoding='utf-8', quotechar='"', escapechar='\\', error_bad_lines=False)
    except Exception as e:
        print(f"CSV 파일 읽기 오류: {e}")
        return []

    # 병렬 처리를 위한 Executor 초기화
    with concurrent.futures.ThreadPoolExecutor() as executor:
        future_to_case = {executor.submit(summarize_case_law, row['판례내용']): row['판례일련번호'] for index, row in df.iterrows()}
        summaries = []

        for future in concurrent.futures.as_completed(future_to_case):
            case_id = future_to_case[future]
            try:
                summary = future.result()
            except Exception as e:
                print(f"요약 생성 오류: {e}")
                summary = None
            summaries.append((case_id, summary))

    return summaries

# 사용 예시
csv_file_path = "/content/NIA2023 법률 텍스트 분석_박재현.csv"  # 여기에 CSV 파일 경로를 입력하세요.
case_summaries = summarize_cases_from_csv(csv_file_path)

for case_id, summary in case_summaries:
    print(f"판례일련번호: {case_id}, 요약: {summary}\n")
```
