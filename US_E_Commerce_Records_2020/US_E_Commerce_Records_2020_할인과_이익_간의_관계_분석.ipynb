```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    data_discount_profit = pd.read_csv('/content/US  E-commerce records 2020.csv', encoding='ISO-8859-1')
except Exception as e:
    error_message = str(e)

# Group by discount and get average profit
grouped_data = data_discount_profit.groupby('Discount')['Profit'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(grouped_data['Discount'], grouped_data['Profit'], marker='o', color='b', linestyle='-')
plt.title('Relationship between Discount and Average Profit')
plt.xlabel('Discount Rate')
plt.ylabel('Average Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```
