```python
import pandas as pd
import matplotlib.pyplot as plt

# Attempt to load the dataset using a different encoding
try:
    data = pd.read_csv("/content/US  E-commerce records 2020.csv", encoding='ISO-8859-1', parse_dates=["Order Date"])
except Exception as e:
    error_message = str(e)

# Check if the data was loaded successfully
if 'data' in locals():
    # Extract month from the "Order Date" and aggregate sales
    monthly_sales = data.groupby(data['Order Date'].dt.month).agg(Total_Sales=('Sales', 'sum'))

    # Plotting the data
    plt.figure(figsize=(12, 6))
    monthly_sales['Total_Sales'].plot()
    plt.title('Monthly Sales Trend')
    plt.xlabel('Month')
    plt.ylabel('Total Sales')
    plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
else:
    print(f"Error encountered: {error_message}")
```
