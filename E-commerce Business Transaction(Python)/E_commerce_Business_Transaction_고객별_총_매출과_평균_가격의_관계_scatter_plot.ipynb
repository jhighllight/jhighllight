```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/customer_stats.csv')

# Scatter plot for Total_Revenue vs Avg_Price
plt.figure(figsize=(10, 6))
plt.scatter(df["Total_Revenue"], df["Avg_Price"], alpha=0.5)
plt.title("Relationship between Total Revenue and Average Price by Customer")
plt.xlabel("Total Revenue")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```
